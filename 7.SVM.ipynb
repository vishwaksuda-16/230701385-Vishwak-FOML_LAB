```python
import cv2, os
import numpy as np
from sklearn import svm
from skimage.feature import hog

# Load face detector
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

data, labels = [], []

# -------- TRAIN --------
for person in os.listdir("C:/Users/USER/FOML/dataset"):
    path = os.path.join("C:/Users/USER/FOML/dataset", person)

    if not os.path.isdir(path):
        continue

    for img_name in os.listdir(path):
        if not img_name.endswith(('.jpg','.png','.jpeg')):
            continue

        img = cv2.imread(os.path.join(path, img_name))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        faces = face_cascade.detectMultiScale(gray, 1.3, 5)

        for (x,y,w,h) in faces:
            face = cv2.resize(gray[y:y+h, x:x+w], (100,100))
            features = hog(face)
            data.append(features)
            labels.append(person)

# Train SVM
model = svm.SVC(kernel='linear')
model.fit(data, labels)

# -------- TEST --------
img = cv2.imread("C:/Users/USER/FOML/test1.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.3, 5)

for (x,y,w,h) in faces:
    face = cv2.resize(gray[y:y+h, x:x+w], (100,100))
    features = hog(face).reshape(1,-1)

    name = model.predict(features)[0]

    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.putText(img,name,(x,y-10),0,0.8,(0,255,0),2)

cv2.imshow("Result", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
