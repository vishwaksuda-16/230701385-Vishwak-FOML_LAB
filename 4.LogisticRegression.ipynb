```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Sample dataset: hours studied vs pass/fail (1 = pass, 0 = fail)
# Example data: [hours studied], [result]
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])   # hours studied
y = np.array([0, 0, 0, 1, 1, 1, 1, 1])                   # pass/fail labels

# Create and train logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Predict for a new student
hours = np.array([[3.5]])   # Example: 3.5 hours of study
prediction = model.predict(hours)
probability = model.predict_proba(hours)

print(f"Hours studied: {hours[0][0]}")
print("Prediction:", "Pass" if prediction[0] == 1 else "Fail")
print(f"Probability of Fail: {probability[0][0]:.2f}, Pass: {probability[0][1]:.2f}")
```
