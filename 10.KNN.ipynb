```python
import math
from collections import Counter

# Function to calculate Euclidean distance
def euclidean_distance(p1, p2):
    sum = 0
    for i in range(len(p1)):
        sum += (p1[i] - p2[i]) ** 2
    return math.sqrt(sum)

# KNN function
def knn(train_data, train_labels, test_point, k):
    distances = []

    # Calculate distance from test point to all training data
    for i in range(len(train_data)):
        dist = euclidean_distance(train_data[i], test_point)
        distances.append((dist, train_labels[i]))

    # Sort distances
    distances.sort()

    # Get k nearest neighbors
    neighbors = distances[:k]

    # Extract labels of neighbors
    labels = [label for (_, label) in neighbors]

    # Majority voting
    most_common = Counter(labels).most_common(1)
    return most_common[0][0]

# Sample dataset
train_data = [
    [2, 4],
    [4, 6],
    [6, 8],
    [8, 10]
]

train_labels = ['A', 'A', 'B', 'B']

# Test point
test_point = [5, 7]

# Value of k
k = 3

# Prediction
result = knn(train_data, train_labels, test_point, k)

print("Test Point:", test_point)
print("Predicted Class:", result)
```
