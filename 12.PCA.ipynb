```python
import numpy as np

# Step 1: Sample dataset (2D → reduce to 1D)
data = np.array([
    [2, 4],
    [4, 6],
    [6, 8],
    [8, 10]
])

# Step 2: Compute mean
mean = np.mean(data, axis=0)

# Step 3: Center the data
centered_data = data - mean

# Step 4: Compute covariance matrix
cov_matrix = np.cov(centered_data.T)

# Step 5: Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 6: Sort eigenvectors by eigenvalues (descending)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvectors = eigenvectors[:, sorted_indices]

# Step 7: Select top k eigenvectors (k = 1)
k = 1
principal_components = eigenvectors[:, :k]

# Step 8: Transform data
reduced_data = np.dot(centered_data, principal_components)

# Output
print("Original Data:\n", data)
print("\nReduced Data (1D):\n", reduced_data)
```
